```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Graph

```python
x  = [1,2,3]
y = [5,6,7]

plt.plot(x,y, label='2x', color='red',linewidth=3, linestyle='--', marker='o', markersize=10, markeredgecolor='blue')

plt.ylabel("Y label",fontdict={'fontsize':20})
plt.xlabel("X label",fontdict={'fontsize':20})

plt.title("First graph", fontdict={"fontname":'arial', 'fontsize':20})

plt.legend()

plt.show()
```

```python
x  = [1,2,3]
y = [4,5,6]

plt.plot(x,y, 'b^--', label='2x')

#line number 2
x2 = np.arange(0,4.5,0.5,)
plt.plot(x2[:6], x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')

plt.title("First graph", fontdict={"fontname":'arial', 'fontsize':20})
plt.ylabel("Y label")
plt.xlabel("X label")


plt.xticks([0,1,2,3,4])
plt.legend()

#resize your graph
plt.figure(figsize=(5,3),dpi=600)

#save
plt.savefig('myfig.png',dpi=300)

plt.show()
```

# barchart

```python
labels = ['A','B','C']
values = [1,2,4]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('-')

plt.figure(figsize=(6,4))
plt.show()
```
